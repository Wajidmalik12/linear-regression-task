```python
import numpy as np
x = np.array([1,3,5])
y = np.array([40000,80000,120000])
#1 lets find slope 
#first we have to find nominator
mean_xy = np.mean(x*y)
mean_x = np.mean(x)
mean_y = np.mean(y)
nominator = mean_xy - (mean_x*mean_y)
#Now we have find denominator
mean_x2 = np.mean(x**2)
squar_mean_x = mean_x**2
denominator = mean_x2 - squar_mean_x
m = nominator / denominator
#2 lets find b 
b = mean_y - m*mean_x
def predected_value(new_x):
    print(m * new_x +b)

predected_value(0)
predected_value(3)
predected_value(7)
predected_value(10)
```
